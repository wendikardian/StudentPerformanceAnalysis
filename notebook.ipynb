## Label Encoder & Scalling

This code defines two functions, scaling and encoding, which use MinMaxScaler and LabelEncoder respectively to standardize numerical and encode categorical features in a dataset, saving the fitted scalers and encoders in a 'model' directory for later use, and ensuring these transformations are consistent across both training and test datasets.

```python
import os
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
import numpy as np
import joblib

# Ensure the 'model' directory exists
os.makedirs('./model', exist_ok=True)

def scaling(features, df, df_test=None):
    if df_test is not None:
        df = df.copy()
        df_test = df_test.copy()
        for feature in features:
            scaler = MinMaxScaler()
            X = np.asanyarray(df[feature])
            X = X.reshape(-1, 1)
            scaler.fit(X)
            df[feature] = scaler.transform(X)
            joblib.dump(scaler, f"./model/scaler_{feature}.joblib")

            X_test = np.asanyarray(df_test[feature])
            X_test = X_test.reshape(-1, 1)
            df_test[feature] = scaler.transform(X_test)
        return df, df_test
    else:
        df = df.copy()
        for feature in features:
            scaler = MinMaxScaler()
            X = np.asanyarray(df[feature])
            X = X.reshape(-1, 1)
            scaler.fit(X)
            df[feature] = scaler.transform(X)
            joblib.dump(scaler, f"./model/scaler_{feature}.joblib")
        return df

def encoding(features, df, df_test=None):
    if df_test is not None:
        df = df.copy()
        df_test = df_test.copy()
        for feature in features:
            encoder = LabelEncoder()
            encoder.fit(df[feature].unique().tolist() + df_test[feature].unique().tolist())  # Fit on combined values
            df[feature] = encoder.transform(df[feature])
            joblib.dump(encoder, f"./model/encoder_{feature}.joblib")

            # Handle unseen values in df_test during transformation
            df_test[feature] = df_test[feature].map(lambda s: encoder.transform([s])[0] if s in encoder.classes_ else -1)
        return df, df_test
    else:
        df = df.copy()
        for feature in features:
            encoder = LabelEncoder()
            encoder.fit(df[feature])
            df[feature] = encoder.transform(df[feature])
            joblib.dump(encoder, f"./model/encoder_{feature}.joblib")
        return df
```

```python
from sklearn.model_selection import GridSearchCV

param_grids = {
    'Logistic Regression': {
        'C': [0.1, 1, 10],
        'solver': ['liblinear', 'saga']
    },
    'Random Forest': {
        'n_estimators': [100, 200],
        'max_depth': [10, 20, None],
        'min_samples_split': [2, 5],
    },
    # 'SVM': {
    #     'C': [0.1, 1, 10],
    #     'kernel': ['linear', 'rbf'],
    #     'gamma': ['scale', 'auto']
    # }
}

grid_search_results = {}
for name, model in models.items():
    print(f"Performing GridSearchCV for {name}...")

    grid_search = GridSearchCV(estimator=model, param_grid=param_grids[name], cv=5, n_jobs=-1, verbose=1)
    grid_search.fit(X_train, y_train)

    best_model = grid_search.best_estimator_
    grid_search_results[name] = {
        'best_params': grid_search.best_params_,
        'best_score': grid_search.best_score_,
        'best_model': best_model
    }

    joblib.dump(best_model, f"model/{name}_best_model.joblib")

    y_pred = best_model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Test accuracy for {name} with best parameters: {accuracy:.4f}")
```

```python

print("Best Parameters for Each Model:")
for model_name, result in grid_search_results.items():
    print(f"{model_name}: Best Parameters = {result['best_params']}")


print("\nBest Cross-Validation Scores for Each Model:")
for model_name, result in grid_search_results.items():
    print(f"{model_name}: Best CV Score = {result['best_score']:.4f}")
```
